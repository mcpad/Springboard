```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_raw = pd.read_csv('../data/SFS_metadata.csv')
data_raw.head(20)
```

```python
data_raw.describe()
```

```python
#remove all nans
data = pd.DataFrame()
data = data_raw.dropna()
data.describe()
```

Convert time to a datetime object
make styles and tags entries lists
strip brand names

```python
data['time'] = data['time'].str.lstrip('Updated on ')
data['time'].head()
```

```python
data['time'] = pd.to_datetime(data['time'])
data['time'].head()
```

```python
styleList = []
for index, row in data.iterrows():
    style = pd.Series(row['styles'].split(','))
    style = style.str.lower()
    style = list(filter(None, style))
    styleList.append(style)
    
styles = pd.DataFrame(styleList)
styles.head()
```

```python
#clean multi-word strings
styleListOneWord = []
for l in styleList:
    row = []
    for item in l:
        if item != None:
            words = item.split()
            if len(words) == 1:
                row.append(words[0])
    styleListOneWord.append(row)
    
styles = pd.DataFrame(styleListOneWord)
styles.head(10)
```

```python
styles.describe()
```
